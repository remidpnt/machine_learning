```python
import fnmatch
from PIL import Image
import os, sys
import torchvision
from torchvision import transforms
import torch
```

```python
i=0
for path, subdirs, files in os.walk("/home/remi/Downloads/data0"):
    for name in files:
        if fnmatch.fnmatch(name, '*.jpg'):
            im = Image.open(os.path.join(path, name))
            imResize = im.resize((64,64), Image.ANTIALIAS)
            imResize.save(os.path.join("/home/remi/Downloads/data64/"+name), 'JPEG', quality=90)
            i+=1
            if i%10000==0:
                print("Preprocess10000")

            
```
